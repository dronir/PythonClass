# Matplotlib

Matplotlib is the _de facto_ plotting library for Python. On one hand, making simple plots is very straightforward. One the other hand, it is a very powerful package, and can make a wide variety of plots. Unfortunately this latter also sometimes hard to use. Its default plots are also not always very good looking and the user must always do some style adjustments. These days there are some libraries which use Matplotlib only as a backend and present the user with more streamlined plot tools.

The `matplotlib` package contains a submodule called `pyplot`, which contains all the main plotting functions that are normally used. It is traditionally imported with the shorter name `plt`.

```python
# this line is only for this web notebook:
%matplotlib inline 

import matplotlib.pyplot as plt
import numpy as np
```

**NOTE**: For most applications, the plots are not drawn until the function `plt.show()` is called, or the plot is written to disk. In this notebook, the plots appear automatically after each code cell ends. Imagine a `plt.show()` at the end of each code cellâ€¦

## Plotting curves

Plotting curves is done with the function `plot`. At its simplest, it takes in one vector or list of values, and plots them against the index.

```python
A = np.random.randn(50)

plt.plot(A)
```

The `plot` function can also take two lists, which are then plotted against each other. A third parameters can be a string which gives the line style.

The `plot` function can also be given many named parameters to control the style, here we use `linewidth` and `color`.

```python
pi = np.pi
X = np.linspace(-2*pi, 2*pi, 100)
Y = np.sin(X)
Z = np.cos(X)

plt.plot(X, Y, "o-", color="black", linewidth=2)
```

Horizontal and vertical lines across the figure can be drawn with `axhline` and `axvline`. Their first parameter is the $x$ or $y$ coordinate of the line, and they can also be given style parameters.

The functions `xlim` and `ylim` are used to set the limits. `title`, `xlabel` and `ylabel` are used to draw a title for the whole figure, as well as the axes.

The `label` parameter to the `plot` function is used to give labels, which are then used by `legend` to draw a legend to the figure.

```python
plt.axhline(0.0, color="black")
plt.axvline(0.0, color="black")

plt.plot(X, Y, "--", color="red", linewidth=3, label="y = sin(x)")
plt.plot(X, Z, ":", color="blue", linewidth=3, label="y = cos(x)")

plt.legend()

plt.ylim(-1.2, 1.2)
plt.xlim(-2*pi, 2*pi)

plt.title("Sinusoids", fontsize="x-large")
plt.xlabel("x", fontsize="x-large")
plt.ylabel("y", fontsize="x-large")

plt.xticks([-2*pi, -pi, 0, pi, 2*pi], 
           ["â€“2Ï€", "â€“Ï€", 0, "Ï€", "2Ï€"], 
           fontsize="x-large")

plt.yticks(fontsize="x-large")
```

## Scatter plots

Scatter plots are made with the `scatter` function. It takes two sets of values to plot against each other. The third parameter sets the size of the dots.

The symbol used for the points can be adjusted in various ways, if needed.

```python
Y2 = 2*X + np.random.randn(100)
Y3 = X + np.random.randn(100) * 2

plt.scatter(X, Y2, 20, color="black", marker="v")
plt.scatter(X, Y3, 30, color="red", marker="*")
```

## Saving plots to disk

Plots can be saved to disk with the `plt.savefig` function. It takes a filename and saves the current figure to that file. It will automatically try to guess the file type based on the name. PDF and PNG files are the most recommended formats.

```python
plt.scatter(X, Y2, 20, color="black", marker="v")
plt.savefig("example.pdf")
```
