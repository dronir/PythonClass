# Seaborn (statistical plots)

Seaborn is a statistical graphics library. It can only create a few different types of plots, but makes creating them very easy.

Seaborn uses Matplotlib in the background, it just replaces the default appearance, and adds functions which automatically make certain complicated plots.

Seaborn is meant to be used with the Pandas library, which provides a data frame type for storing statistical data with e.g. categorical fields. Many of the plots can made automatically from Pandas data.

```python
%matplotlib inline

import numpy as np

import seaborn as sns
```

```python
X = np.linspace(0, 5, 10)
Y1 = 2*X - 2 + np.random.randn(10)
Y2 = -1*X + 0.5 + np.random.randn(10)
```

## Regression line plots

Here we make automatically a plot which contains a scatterplot of the data points, together with a best-fit line and the error envelope for the line.

```python
sns.set_style("darkgrid")

sns.regplot(x=X, y=Y1)
sns.regplot(x=X, y=Y2)
```

## Re-styling Matplotlib

When Seaborn is imported, its graphical style applies to all of Matplotlib. Using Matplotlib in the regular way after Seaborn was impored makes all the Matplotlib plots look like Seaborn.

```python
import matplotlib.pyplot as plt

plt.scatter(X, Y1)
plt.scatter(X, Y2)
```
